```python
from IPython.display import clear_output
import random


class player:
       def __init__(self,name,symbol,turn = False):
            self.name = name
            self.symbol = symbol
            self.turn = turn

class HumanPlayer(player):
    def __init__(self,name,symbol,turn = False):
        player.__init__(self,name,symbol,turn)

class ComputerPlayer(player):
    def __init__(self,name,symbol,turn,smart):
        player.__init__(self,name,symbol,turn)
        self.smart_Bot = smart
    

class tic_tac_toe:
     
    def __init__(self):
        self.setup()
     
    def setup(self):
        self.game_over = False
        
        self.board = [[" "," "," "],
                     [" "," "," "],
                     [" "," "," "]]
        
        self.winner = None
        
    def display(self,player1,player2):
        print(f'{player1.name} = {player1.symbol} vs {player2.name} = {player2.symbol}\n')
        print(f"{' | '.join(self.board[0])}\n--+---+--\n{' | '.join(self.board[1])}\n--+---+--\n{' | '.join(self.board[2])}")
    
    def check_game_status(self):
        if(self.game_over == False):
            self.check_rows()
        if(self.game_over == False):
            self.check_columns()
        if(self.game_over == False):
            self.check_diagonal()
        if(self.game_over == False):
            self.check_full()
    
    def check_full(self):
        if (" " not in self.board[0]) and (" " not in self.board[1]) and (" " not in self.board[2]):
            self.game_over = True
            
    def check_rows(self):
        for row in range(3):
            if (self.board[row][0] == self.board[row][1]) and (self.board[row][0] == self.board[row][2]) and (self.board[row][0] != " "):
                self.winner = self.board[row][0]
                self.game_over = True
                break
                  
    def check_columns(self):
        for col in range(3):
            if (self.board[0][col] == self.board[1][col]) and (self.board[0][col] == self.board[2][col]) and (self.board[0][col] != " "):
                self.winner = self.board[0][col]
                self.game_over = True
                break
    def check_diagonal(self):
        if (self.board[0][0] == self.board[1][1]) and (self.board[0][0] == self.board[2][2]) and (self.board[0][0] != " "):
            self.winner = self.board[1][1]
            self.game_over = True
        elif (self.board[0][2] ==self.board[1][1]) and  (self.board[0][2] == self.board[2][0]) and (self.board[0][2] != " "):
            self.winner = self.board[1][1]
            self.game_over = True
    

            

        
    #human vs human    
    def human_vs_human(self):
        self.setup()
        name = input("Enter player 1 name: ")
        player_1 = HumanPlayer(name,'X',turn = True)
        name = input("Enter player 2 name: ")
        player_2 = HumanPlayer(name,'O')
        current_player = player_1
               
        while(not self.game_over):
            clear_output()
            self.display(player_1,player_2)
            self.play(current_player)
            self.check_game_status()
            current_player = self.flip_player(player_1,player_2)
        clear_output()
        self.display(player_1,player_2)
        if(self.winner != None):
            winner = player_1 if player_1.symbol == self.winner else player_2
            print(f"\nGAME OVER - the winner is: {winner.name} use {winner.symbol}")
        else:
            print('\nGAME OVER - TIE')
    
    
    #human vs dummy bot (human start first)
    def human_vs_dummy_bot(self):
        self.setup()
        name = input("Enter player 1 name: ")
        player_1 = HumanPlayer(name,'X',turn = True)
        player_2 = ComputerPlayer("DummyBot",'O',turn= False, smart = False)  
        current_player = player_1
        
        while(not self.game_over):
            clear_output()
            self.display(player_1,player_2)
            self.play(current_player)
            self.check_game_status()
            current_player = self.flip_player(player_1,player_2)
        clear_output()
        self.display(player_1,player_2)
        if(self.winner != None):
            winner = player_1 if player_1.symbol == self.winner else player_2
            print(f"\nGAME OVER - the winner is: {winner.name} use {winner.symbol}")
        else:
            print('\nGAME OVER - TIE')
            
   

    #human vs smart bot (bot start first)
    def human_vs_smart_bot(self):
        self.setup()
        name = input("Enter player 1 name: ")
        
        player_1 = HumanPlayer(name,'O',turn = False)
        player_2 = ComputerPlayer("Smart Bot",'X',turn = True,smart = True)
        
        current_player = player_2
        
        while(not self.game_over):
            #clear_output()
            self.display(player_1,player_2)
            self.play(current_player)
            self.check_game_status()
            current_player = self.flip_player(player_1,player_2)
        #clear_output()
        self.display(player_1,player_2)
        if(self.winner != None):
            winner = player_1 if player_1.symbol == self.winner else player_2
            print(f"\nGAME OVER - the winner is: {winner.name} use {winner.symbol}")
        else:
            print('\nGAME OVER - TIE')
            
    def is_valid_position(self,position,player):
        if isinstance(player,HumanPlayer):
            if( position >= 1 and position <= 9 ):
                col = (2 if position % 3 == 0 else position % 3 - 1)
                row = (position // 3 - 1 if position % 3 == 0 else position // 3)     
                if self.board[row][col] == " ":
                    return True
        elif self.board[position[0]][position[1]] == " ":
            return True
        
        return False
    
    def get_available_moves(self):
        lst = []
        for row_index,row in enumerate(self.board):
            for col_index,item in enumerate(row):
                if self.board[row_index][col_index] == " ":
                    lst.append((row_index,col_index))
        return lst
    
    def next(self,symbol):
        return ('O' if symbol == 'X' else 'X')
    
    def get_score(self,is_max):      
        if self.winner == None:
            return 0
        if is_max == True:
            return 1 
        return -1
    
    def finish(self):
        self.check_game_status()
        if self.game_over == True:
            self.game_over = False
            return True
        return False
   
    #need to debug this code - something wrong with algorithm - should be smarter
    def minimax_algorithm(self,is_maximizing_player):       

        if self.finish() == True:
            score = self.get_score(is_maximizing_player)
            self.winner = None
            return score
      
        elif is_maximizing_player == True:  #maximize player
            
            best_score = float('-inf')
            
            for row_index,row in enumerate(self.board):
                for col_index,item in enumerate(row):
                    
                    if self.board[row_index][col_index] == " ":
                        self.board[row_index][col_index] = 'X'                    
                        score = self.minimax_algorithm(not(is_maximizing_player))
                        self.game_over = False
                        self.winner = None
                        self.board[row_index][col_index] = " "
                        best_score = max(score,best_score)
         
            return best_score
        
        elif is_maximizing_player == False: #minimize player
            
            best_score = float('inf')
            
            for row_index,row in enumerate(self.board):
                for col_index,item in enumerate(row):
                    
                    if self.board[row_index][col_index] == " ":
                        self.board[row_index][col_index] = 'O'                 
                        score = self.minimax_algorithm(not(is_maximizing_player))
                        self.game_over = False
                        self.winner = None
                        self.board[row_index][col_index] = " "
                        best_score= min(score,best_score)
             
            return best_score
   
    #TODO - cleaner code but still have bug
    def minimax_algorithm2(self,is_maximizing_player):       
       
        print(f' is_maximizing_player: {is_maximizing_player} ')
        if self.finish() == True:
            score = self.get_score(is_maximizing_player)
            self.winner = None
            return score
        else:
            
            best_score = float('-inf') if is_maximizing_player == True else float('inf')
            
            for row_index,row in enumerate(self.board):
                for col_index,item in enumerate(row):
                    
                    if self.board[row_index][col_index] == " ":
                        print(f' - [{row_index},{col_index} = {is_maximizing_player}]',end = " ")
                        self.board[row_index][col_index] = 'X' if is_maximizing_player == True else 'O'                    
                        score = self.minimax_algorithm(not(is_maximizing_player))
                        self.game_over = False
                        self.winner = None
                        self.board[row_index][col_index] = " "
                        best_score = max(score,best_score) if is_maximizing_player == True else min(score,best_score)
         
            return best_score
        
       
    
    def get_ideal_move(self,player):
        
        best_score = float('-inf')
        best_move = []
        
        for row_index,row in enumerate(self.board):
            for col_index,item in enumerate(row):
                if self.board[row_index][col_index] == " ":
                    self.board[row_index][col_index] = player.symbol
                    score = self.minimax_algorithm(False)
                    self.game_over = False
                    self.winner = None
                    self.board[row_index][col_index] = " "
                    if(score > best_score):
                        best_score = score
                        best_move = [row_index,col_index]
        return best_move

    def set_move(self,curr_player):      
        if isinstance(curr_player,HumanPlayer): 
            while(True):
                try:
                    position = int(input(f"[{curr_player.name}] Enter position 1-9: "))
                    break
                except:
                    print("wrong input\n")
        
        elif isinstance(curr_player,ComputerPlayer) and curr_player.smart_Bot == False:
            avilable_moves = self.get_available_moves()
            position = random.choice(avilable_moves)
       
        elif isinstance(curr_player,ComputerPlayer) and curr_player.smart_Bot == True:
            position = self.get_ideal_move(curr_player)
            print(position)
            
        return position
            
        
    def play(self,curr_player):
        while(True):
          
                position = self.set_move(curr_player)
            
                if( self.is_valid_position(position,curr_player) ):
                    if isinstance(curr_player,HumanPlayer):
                        col = (2 if position % 3 == 0 else position % 3 - 1)
                        row = (position // 3 - 1 if position % 3 == 0 else position // 3)     
                    elif isinstance(curr_player,ComputerPlayer):
                        col = position[1]
                        row = position[0]
                    
                    self.board[row][col] = curr_player.symbol
                    break
                print("wrong position\n")
       
    def flip_player(self,player1,player2):
        if(player1.turn == True):
            player1.turn = False
            player2.turn = True
            return player2
        else:
            player1.turn = True
            player2.turn = False
            return player1
            
```
